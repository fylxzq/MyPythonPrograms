# 5K-Means

```python
df = spark.read.format("text").option("header","true").load("inputData/protein.txt",sep="\t")
```

```python
import pandas as pd
```

```python
data = pd.read_table("inputData/protein.txt",sep="\t")
```

```python
df = spark.createDataFrame(data)
```

```python
df.printSchema()
train_df,test_df = df.randomSplit([0.7,0.3])
```

```python
from pyspark.ml.feature import Normalizer
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.clustering import KMeans
from pyspark.ml import Pipeline
assemblerInput = df.columns[1:-1]
assembler = VectorAssembler(inputCols=assemblerInput,outputCol="features")
normal = Normalizer(inputCol="features",outputCol="nor-features")
```

```python
model = KMeans(featuresCol="nor-features",initSteps=10,maxIter=200,predictionCol="rst",k=5)
```

```python
pipeline = Pipeline(stages=[assembler,normal,model])
```

```python
pipelineModel = pipeline.fit(df)
```

```python
prediction = pipelineModel.transform(df)
```

```python
prediction_pandas = prediction.toPandas()
prediction_pandas
```

```python
rst = prediction_pandas["rst"]
```

```python
rst = list(rst)
```

```python
rst
```

```python
def print_kmcluster(k):
    for i in range(k):
        print("聚类",i)
        ls = []
        for index,value in enumerate(rst):
            if i == value:
                ls.append(index)
        print(prediction_pandas.loc[ls,["Country","RedMeat","Fish","Fr&Veg"]])
print_kmcluster(5)
```

# 6.随机森林回归

```python
import pandas as pd
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
x = boston.data
y = boston.target
```

```python
df = pd.DataFrame(x)
```

```python
df["13"] = pd.Series(y)
```

```python
df1= spark.createDataFrame(df)
```

```python
df2 = df1.select(df1.columns[:-2]+[df1["13"].alias("label")])
train_df,test_df=df2.randomSplit([0.7,0.3])
```

```python
df2.show()
```

```python
from pyspark.ml.feature import VectorAssembler,StandardScaler
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml import Pipeline
```

```python
assemblerInput = df2.columns[:-2]
assembler = VectorAssembler(inputCols=assemblerInput,outputCol="features")
scaler = StandardScaler(inputCol="features",outputCol="scaler_features")
model = RandomForestRegressor(featuresCol="scaler_features",labelCol="label")
pipeline = Pipeline(stages=[assembler,scaler,model])
```

```python
pipelineModel = pipeline.fit(train_df)
```

```python
prediction = pipelineModel.transform(test_df)
```

```python
prediction.show()
```

```python
from pyspark.ml.evaluation import RegressionEvaluator
```

```python
evalutor = RegressionEvaluator(predictionCol="prediction",labelCol="label",metricName="rmse")
```

```python
rmse = evalutor.evaluate(prediction)
```

```python
evalutor1 = RegressionEvaluator(predictionCol="prediction",labelCol="label",metricName="mse")
mse = evalutor1.evaluate(prediction)
print("均方误差为",mse)
```

```python
print("均方根误差为",rmse)
```

```python
evalutor2 = RegressionEvaluator(predictionCol="prediction",labelCol="label",metricName="mae")
mae = evalutor2.evaluate(prediction)
print("均方误差为",mae)
```
