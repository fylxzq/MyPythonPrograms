```python
team_data= team_data.loc[team_data['team_name']!='']
```

```python
lake_data = pd.read_csv('inputData/lakes_data.csv')
```

```python
lake_data.head()
```

```python
def get_season(game_time):
    ymd_lst = game_time.split('-')
    year = ymd_lst[0]
    month = int(ymd_lst[1])
    day = ymd_lst[2]
    if(month>9):
        return int(year)
    else:
        return int(year) -1
```

```python
lake_data['season'] = lake_data['game_time'].apply(get_season)
```

```python
lake_data.head()
```

```python
lake_data['game_time'] = pd.to_datetime(lake_data['game_time']).map(lambda x:str(x.month)+'-'+str(x.day))
```

```python
lake_data.head()
```

```python
lake_data['rival'] = lake_data['rival'].str.extract('(\d{0,2}.*?)\d')
```

```python
lake_data = lake_data.loc[lake_data['rival']!=' ']
```

```python
teams = lake_data['rival'].unique()
teams
```

```python
def get_team(x):
    for i in teams:
        if(i in x):
            return i
```

```python
team_data['team_name'].apply(get_team)
```
