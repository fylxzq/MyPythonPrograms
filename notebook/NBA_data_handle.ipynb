```python
import pandas as pd
```

```python
game_records = pd.read_csv('inputData/nba_races.csv')#读取常规赛历史比赛数据
teams_info = pd.read_csv('inputData/team_data2.csv').drop(['coach_off_times','coach_wins_rate','coach_champions'],axis=1)#读取球队历史战绩数据
game_records['label'] = (game_records.hostscore > game_records.visitscore)#生成标签字段
game_records['label'] = [1 if i==True else 0 for i in game_records['label']]
game_records['year'] = game_records['game_time'].map(lambda x:int(x[:4]) if int(x[5:7]) > 8 else int(x[:4])-1)#读取赛季年份，用于数据连接
hostteam_info = teams_info.rename(columns={'team_name':'hostteam','win_rates':'h_rates'}).drop(['wins','loss'],axis=1)#主场球队战绩信息
visitteam_info = teams_info.rename(columns={'team_name':'visitteam','win_rates':'v_rates'}).drop(['wins','loss'],axis=1)#客场球队战绩
game_records = pd.merge(game_records,visitteam_info,on=['year','visitteam'])#连接
game_records = pd.merge(game_records,hostteam_info,on=['year','hostteam'])#l连接
stars = pd.read_csv('inputData/stars.csv')#读取全明星数据
star1 = stars.rename(columns={'team_name':'hostteam','stat_num':'h_stars'})#连接
star2 = stars.rename(columns={'team_name':'visitteam','stat_num':'v_stars'})#连接
game_records = pd.merge(star1,game_records,on=['year','hostteam'])
game_records = pd.merge(star2,game_records,on=['year','visitteam'])
```

```python
game_records['h_rates'] = game_records['h_rates'].str.strip('%').astype(float)#去除百分号
```

```python
game_records['v_rates'] = game_records['v_rates'].str.strip('%').astype(float)
```

```python
game_records['game_time'] = game_records['game_time'].map(lambda x:int(x[5:7]))
```
