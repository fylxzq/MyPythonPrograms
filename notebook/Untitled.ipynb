# 机器学习

```python
import pandas as pd
```

```python
team_info = pd.read_csv('inputData/evl_team_score.csv')
```

```python
team_info
```

```python
team_info['t1'] = team_info['Tm_FGA']+0.4*team_info['Tm_FTA']-1.07*(team_info['Tm_ORB'] / (team_info['Tm_ORB']+team_info['Opp_DRB']))*(team_info['Tm_FGA']-team_info['Tm_FG'])+team_info['Tm_TOV']
```

```python
team_info['t2'] = team_info['Opp_FGA']+0.4*team_info['Opp_FTA']-1.07*(team_info['Opp_ORB'] / (team_info['Opp_ORB']+team_info['Tm_DRB']))*(team_info['Opp_FGA']-team_info['Opp_FG'])+team_info['Opp_TOV']
```

```python
team_info['score'] = (team_info['t1']+team_info['t2'])*0.5
```

```python
tmp_df = team_info[['team_name','score']].sort_values('score',ascending=False)
```

```python
team_name = tmp_df['team_name']
```
