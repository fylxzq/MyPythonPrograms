```python
import numpy as np
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

a = np.arange(0.0,5.0,0.2)

plt.subplot(211)
plt.plot(a,f(a))

plt.subplot(212)
plt.plot(a,np.cos(2*np.pi*a),'r--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as pla
from  scipy.io import wavfile

rate_h,hstrain = wavfile.read("inputData/H1_Strain.wav","rb")
rate_l,lstrain = wavfile.read("inputData/L1_Strain.wav","rb")
reftime,ref_H1 = np.genfromtxt("inputData/wf_template.txt").transpose()

htime_interval = 1/rate_h
ltime_interval = 1/rate_l

htime_len = hstrain.shape(0)/rate_h
htime = np.arange(-htime_len/2,htime_len/2,htime_interval)
ltime_len = lstrain.shape(0)/rate_l
ltime = np.arange(-ltime_len/2,ltime_len/2,ltime_interval)

fig = plt.figure(figsize=(12,6))

plth = fig.add_subplot(221)
plth.plot(htime,hstrain,'y')
plth.set_xlabel("Time(Second)")
plth.set_ylabe;("H1 Strain")
plth.set_title("H1 Strain")

pltl = fig.add_subplot(222)
pltl.plot(ltime,lstrain,'g')
pltl.set_xlabel("Time(Second)")
pltl.set_ylabe;("L1 Strain")
pltl.set_title("L1 Strain")

pltref = fig.add_subplot(212)
pltref.plot(reftime,ref_H1)
pltref.set_xlabel("Time(Second)")
pltref.set_ylabe;("Template Strain")
pltref.set_title("Template")
fig.tight_layout()
```
