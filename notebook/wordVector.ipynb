```python
import jieba
import os
import numpy as np
```

```python
txtnameLst = os.listdir("inputData/jiebaTest")
```

```python
raw_data = []
for i in txtnameLst[0:-1]:
    with open("inputData/jiebaTest/"+i,"r") as file:
        raw_data.append(file.readlines())
```

```python
raw_data[0]
```

```python
cut_lst = []
for i in raw_data:
    lst1 = []
    for word in i:
        lst1.append(jieba.cut(word))
    cut_lst.append(lst1)
```

```python
cut_lst[0]
```

```python
words_lst = []
for i in cut_lst:
    lst = []
    for gtr in i:
        for word in gtr:
            lst.append(word)
    words_lst.append(lst)
```

```python
stopwordsLst = []
with open("inputData/jiebaTest/StopList.txt","r") as file:
    for i in file.readlines():
        stopwordsLst.append(i.strip("\n"))
stopwordsLst.append("\n")
```

```python
words_stop_lst = []
for i in words_lst:
    lst1=[]
    for j in i:
        if(j not in stopwordsLst):
            lst1.append(j)
    words_stop_lst.append(lst1)
```

```python
word_dict_lst = []
for i in words_stop_lst:
    dct = {}
    for word in i:
        if(word not in dct):
            dct[word] = 1
        else:
            dct[word]+=1
    word_dict_lst.append(dct)
```

```python
word_dict_lst
```

```python
word_tf_lst = []
for i in word_dict_lst:
    lst = []
    for key,value in i.items():
        lst.append((key,int(value)/max(i.values())))
    word_tf_lst.append(lst)
```

```python
def countnumber(word):
    number = 0
    for i in words_stop_lst:
        if(word in i):
            number +=1
    return number
```

```python
word_idf_lst = []
for i in word_dict_lst:
    lst = []
    for key,value in i.items():
        lst.append((key,np.log(10/(countnumber(key)+1))))
    word_idf_lst.append(lst)
```

```python
word_idf_lst
```

```python
rst_lst = []
for i in range(len(word_tf_lst)):
    lst = []
    for j in range(len(word_tf_lst[i])):
        lst.append((word_idf_lst[i][j][0],word_tf_lst[i][j][1]*word_idf_lst[i][j][1]))
    rst_lst.append(lst)
```
